# Introducción

El Análisis de sentimientos se utiliza para descubrir las opiniones, emociones y sentimientos de las personas sobre un producto o servicio.

En teoría es un estudio estadístico de opiniones, sentimientos, actitudes, puntos de vista, emociones, etc. expresados en texto. Este texto puede estar en una variedad de formatos como Revisiones, Blogs, Noticias o Comentarios.


La capacidad de extraer información de este tipo de datos es una práctica ampliamente adoptada por muchas organizaciones en todo el mundo. Su aplicación es amplia y potente.


El análisis de sentimientos tiene muchos nombres: *minería de opinión*, *minería de sentimientos* y análisis de subjetividad*.

¡Son muchos diferente nombres!


Lo usamos para muchas tareas diferentes, por ejemplo:

- ¿Vale la pena ver esa película?
- ¿Qué piensa la gente sobre el último iPhone?
- ¿Qué piensa la gente sobre las elecciones, los candidatos particulares o los problemas?


Si podemos medir bien estos sentimientos, podemos usarlos para hacer pronósticos interesantes sobre una variedad de cosas, como las tendencias del mercado de valores, los resultados electorales, etc.

## Tipificación de los problemas


La aplicación de análisis de sentimientos es amplia. Algunos problemas son relativamente fáciles y otros son más complejos.

Podemos dividir ampliamente los problemas en el análisis de sentimientos en 3 grandes categorías:

1. **La polaridad del texto** se refiere a la identificación de la orientación del sentimiento del texto
2. **La calificación de sentimiento** da un rango numérico, digamos de 1 a 5 para texto
3. **El análisis de sentimientos basado en aspectos** determina el sentimiento hacia aspectos específicos en el texto.

Sin embargo, dado que la aplicación de análisis de sentimientos es amplia, algunos problemas son fáciles, mientras que otros son más complejos.

## Polaridad

Aquí estamos más interesados en la orientación general del texto, es decir, si es positivo o negativo. 

Debido a la estructura libre del texto, hay muchos casos que son difíciles de resolver.

Por ejemplo, ¿es esta una crítica positiva o negativa?

- *Fácil*: “Compré el iPhone XR hace unos días. Es un teléfono tan bonito, aunque un poco grande y caro. ¡La cámara es increíble!
- *Complicada*: "Honda Accords y Toyota Camrys son buenos sedanes, pero difícilmente el mejor auto en la carretera"

## Califique


Clasificar el sentimiento del texto proporciona más granularidad. Aquí, quiero asignar una puntuación numérica al sentimiento dentro de un rango.

Por ejemplo, podemos tener un modelo que asigne una calificación a las reseñas de Amazon, de 1 a 5, siendo 1 muy negativo y 5 muy positivo.

## Métodos para hacer análisis de sentimientos

Puede utilizar tanto el aprendizaje supervisado como los métodos de aprendizaje no supervisados ​​para su análisis de sentimientos.


### Análisis de sentimientos utilizando aprendizaje supervisado


El aprendizaje supervisado para un modelo de sentimiento puede considerarse como el entrenamiento de un clasificador.

El análisis de sentimiento basado en el aprendizaje supervisado consta de dos pasos, el primero es aprender o entrenar y el segundo es probar.



### Análisis de sentimientos utilizando aprendizaje no supervisado


Si bien los métodos de aprendizaje automático se utilizan ampliamente en el análisis de sentimientos, también hay muchos sistemas que adoptan métodos de aprendizaje no supervisados. 

Estos métodos requieren diccionarios de opinión.Hay dos métodos que puede usar para crear un diccionario de sentimientos. 

Un enfoque se basa en fuentes curadas por humanos. Por ejemplo, tenemos diccionarios de opinión como SentiWordnet y el léxico de subjetividad. 

O puede utilizar un enfoque basado en corpus donde se crean diccionarios especializados de sentimientos a partir del corpus de texto.

Un paso intermedio para este tipo de aprendizaje es hacer desarrollar previamente auto-codificadores, un tipo de red neuronal que codifica los datos (en este caso los textos inmersos en una codificación numérica adecuada) y luego decodificarlos. Los datos codificados son clasificados por un algortimo de clasificación automática (aprendizaje semi-supervisado)

Con los léxicos, se utilizan algoritmos de agrupamiento o puntuación para determinar el sentimiento del texto de entrada. Los sistemas de análisis de sentimientos basados en léxico son costosos y difíciles de desarrollar.

Los recursos léxicos de alta calidad son la clave del buen rendimiento. Por lo general, no dan un significado dependiente del dominio o del contexto. 

Para superar estos problemas, utilice un enfoque basado en corpus. En general, los enfoques basados en corpus se basan en patrones sintácticos o de ocurrencia simultánea en grandes corpus. También necesitará un corpus grande para obtener una buena cobertura.

## Algunos consejos útiles

### Aplicar métodos lingüísticos y estadísticos a la tarea de análisis.

El significado varía según el sentido de las palabras, el contexto y lo que se está discutiendo. Necesitamos buscar métodos que apliquen métodos lingüísticos y estadísticos a la tarea de análisis. El análisis de sentimientos a través de la búsqueda de términos en un léxico es un método fácil pero crudo.

### Entrenar modelos adaptados de dominio

Ya sea que aplique ingeniería del lenguaje, métodos estadísticos o aprendizaje automático a la tarea, los modelos adaptados al dominio debidamente entrenados superarán la clasificación genérica.

### Análisis de sentimientos basado en aspectos

El análisis del sentimiento a nivel de documento es en gran medida pasado. Apunte a la resolución de sentimientos a nivel de entidad, concepto o tema.

### Otras fuentes de análisis de sentimientos

El texto es la fuente de datos de opinión más común, pero no es la única. La biometría, las imágenes y el sonido también son fuentes de datos de opinión.

### Computación y Estadística afectiva

En términos más amplios, el Análisis de sentimientos es parte del mundo de la computación y estadística afectiva. La computación afectiva es el estudio y desarrollo de modelos (Estadística) sistemas y dispositivos (computación)  que pueden reconocer, interpretar, procesar y simular los efectos humanos. 

### ¿Es exacto?

Debemos tener cuidado con las afirmaciones de precisión. La mayor amenaza para la precisión en el análisis de sentimientos hoy es la concordancia humana: este es el grado de acuerdo entre los humanos (o entre humanos y máquinas). Numerosos estudios han demostrado que la tasa de concordancia humana está entre 70% y 80%. Hay muchos factores que afectan el rendimiento de un sistema de análisis de sentimientos.

### Omitimos los tweets clasificados como neutral

```python
tweet_df = tweet_df[tweet_df['airline_sentiment'] != 'neutral']
tweet_df.head(10)
```

```python
# convert airline_seentiment to numeric
sentiment_label = tweet_df.airline_sentiment.factorize() # convierte a factores numéricos
sentiment_label
```

## Modelo Predictivo: polaridad

### Preprocesamiento de textos

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tweet = tweet_df.text.values
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(tweet)
vocab_size = len(tokenizer.word_index) + 1
encoded_docs = tokenizer.texts_to_sequences(tweet)
padded_sequence = pad_sequences(encoded_docs, maxlen=200)
```

```python
#print(tokenizer.word_index)
```

```python
print(tweet[0])
print(encoded_docs[0])
```

```python
print(padded_sequence[0])
```

### El Modelo keras

```python
# Build the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense, Dropout, SpatialDropout1D
from tensorflow.keras.layers import Embedding

embedding_vector_length = 32
model = Sequential() 
model.add(Embedding(vocab_size, embedding_vector_length, input_length=200) )
model.add(SpatialDropout1D(0.25))
model.add(LSTM(50, dropout=0.5, recurrent_dropout=0.5))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid')) 
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])  
print(model.summary()) 
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='../Imagenes/model.png')
```

```python
history = model.fit(padded_sequence,sentiment_label[0],validation_split=0.2, epochs=5, batch_size=32)
```

```python
from matplotlib import pyplot as plt

plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```
